```python
import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0}
]

# Step 1 - Deep copy and proof
inventory_backup = copy.deepcopy(inventory)

inventory["Paneer Tikka"]["stock"] = 999
print("inventory stock        :", inventory["Paneer Tikka"]["stock"])
print("inventory_backup stock :", inventory_backup["Paneer Tikka"]["stock"])

inventory["Paneer Tikka"]["stock"] = 10
print("Restored inventory stock:", inventory["Paneer Tikka"]["stock"])

# Step 2 - Deduct cart from inventory
for entry in cart:
    item     = entry["item"]
    quantity = entry["quantity"]
    stock    = inventory[item]["stock"]

    if stock < quantity:
        print("Warning: Not enough stock for", item)
        inventory[item]["stock"] = 0
    else:
        inventory[item]["stock"] = stock - quantity
        print("Deducted", quantity, "of", item, "remaining stock:", inventory[item]["stock"])

# Step 3 - Reorder alerts
print("Reorder Alerts:")
for item in inventory:
    stock         = inventory[item]["stock"]
    reorder_level = inventory[item]["reorder_level"]
    if stock <= reorder_level:
        print("⚠ Reorder Alert:", item, "— Only", stock, "unit(s) left (reorder level:", str(reorder_level) + ")")

# Step 4 - Compare inventory vs backup
print("Final Inventory (after deductions):")
for item in inventory:
    print(" ", item, "- stock:", inventory[item]["stock"])

print()
print("Inventory Backup (unchanged):")
for item in inventory_backup:
    print(" ", item, "- stock:", inventory_backup[item]["stock"])
```

```python
sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

# Step 1 - Revenue per day
print("Revenue Per Day:")
print("--------------------------------")
for date in sales_log:
    day_total = 0
    for order in sales_log[date]:
        day_total = day_total + order["total"]
    print(date, ":", "₹" + str(format(day_total, ".2f")))

print()

# Step 2 - Best selling day
best_day     = ""
best_revenue = 0
for date in sales_log:
    day_total = 0
    for order in sales_log[date]:
        day_total = day_total + order["total"]
    if day_total > best_revenue:
        best_revenue = day_total
        best_day     = date

print("Best Selling Day:", best_day, "-", "₹" + str(format(best_revenue, ".2f")))
print()

# Step 3 - Most ordered item
item_count = {}
for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 0
            item_count[item] = item_count[item] + 1

most_ordered       = ""
most_ordered_count = 0
for item in item_count:
    if item_count[item] > most_ordered_count:
        most_ordered_count = item_count[item]
        most_ordered       = item

print("Most Ordered Item:", most_ordered, "-", most_ordered_count, "orders")
print()

# Step 4 - Add new day and reprint
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("Updated Revenue Per Day:")
print("--------------------------------")
for date in sales_log:
    day_total = 0
    for order in sales_log[date]:
        day_total = day_total + order["total"]
    print(date, ":", "₹" + str(format(day_total, ".2f")))

print()

best_day     = ""
best_revenue = 0
for date in sales_log:
    day_total = 0
    for order in sales_log[date]:
        day_total = day_total + order["total"]
    if day_total > best_revenue:
        best_revenue = day_total
        best_day     = date

print("Updated Best Selling Day:", best_day, "-", "₹" + str(format(best_revenue, ".2f")))
print()

# Step 5 - Numbered list using enumerate
print("All Orders:")
print("--------------------------------")

all_orders = []
for date in sales_log:
    for order in sales_log[date]:
        all_orders.append((date, order))

for i, (date, order) in enumerate(all_orders):
    items_str = ", ".join(order["items"])
    print(str(i + 1) + ".", "[" + date + "]", "Order #" + str(order["order_id"]), "—", "₹" + str(format(order["total"], ".2f")), "— Items:", items_str)
```
